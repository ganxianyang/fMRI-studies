```python
# to run this code, please use Jupyter Notebook

# modified from: https://towardsdatascience.com/make-beautiful-nightingale-rose-chart-in-python-covid-19-death-rate-141a7a49bb61

import pandas as pd
import numpy as np
from pyecharts.charts import Pie
from pyecharts import options as opts

data=pd.read_csv('top15regionsforVIDS_FDR005.CSV')

c = data['subreions_from_brainnetome'].values.tolist()
d = data['VIDS_positive'].values.tolist()

#create the color_series for the rosechart
color_series = ['#9E3A26','#B24223','#C64A22','#CF4F22','#D75521',
 '#DE5D1F','#E4641E','#F0751E','#F28026','#F38B2F',
 '#F49538','#F69E41','#F7A84A','#F8BC58','#F4D166']
rosechart = Pie(init_opts=opts.InitOpts(width='1350px', height='750px'))


# set the color
rosechart.set_colors(color_series)
# add the data to the rosechart
rosechart.add("", [list(z) for z in zip(c, d)],
        radius=["20%", "95%"],  # 20% inside radiusï¼Œ95% ourside radius
        center=["30%", "60%"],   # center of the chart
        rosetype="area",
        label_line_opts = False) # do not show labels surrounding the rosechart)

rosechart.render_notebook()
```
